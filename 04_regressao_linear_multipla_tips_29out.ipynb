```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Carregar dataset
df = sns.load_dataset("tips")

# Transformar variável categórica em numérica
df["smoker"] = df["smoker"].map({"No": 0, "Yes": 1})

# Definir variáveis independentes e dependente
X = df[["total_bill", "size", "smoker"]]
y = df["tip"]

# Criar e treinar modelo
modelo_multi = LinearRegression()
modelo_multi.fit(X, y)

# Previsões
df["pred_tip_multi"] = modelo_multi.predict(X)

# Resultados
print("Intercepto (a):", modelo_multi.intercept_)
for nome, coef in zip(X.columns, modelo_multi.coef_):
    print(f"{nome}: {coef:.3f}")

r2 = r2_score(y, df["pred_tip_multi"])
print(f"\nR² do modelo: {r2:.3f}")

# Visualização: real x previsto
plt.scatter(df["total_bill"], df["tip"], label="Real", alpha=0.6)
plt.scatter(df["total_bill"], df["pred_tip_multi"], color="red", label="Previsto", alpha=0.6)
plt.title("Regressão Linear Múltipla - Gorjeta vs Total da Conta")
plt.xlabel("Total da Conta")
plt.ylabel("Gorjeta")
plt.legend()
plt.show()
```
