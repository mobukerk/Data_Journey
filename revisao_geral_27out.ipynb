```python
import pandas as pd

# Criar DataFrame manual
dados= {
    "Dia":["Seg","Ter","Qua","Qui","Sex"],
    "Horas_Estudo":[2,3,1,4,2],
    "Produtividade":[6.3,7.1,5.9,8.4,6.1]
}

df=pd.DataFrame(dados)
display(df)
print("\nEstatistica basica:\n")
print(df.describe())

# Nova coluna e ordenação
df["Eficiência"] = df["Produtividade"] / df["Horas_Estudo"]
display(df.sort_values("Eficiência", ascending=False))
```

```python
# --- Prática: dataset real ---
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar dados reais
df = sns.load_dataset("tips")

# Mostrar primeiras linhas
display(df.head())

# Estatísticas gerais
print(df.describe())

# Criar coluna de % de gorjeta
df["perc_tip"] = (df["tip"] / df["total_bill"]) * 100

# Visualizar distribuição
sns.histplot(df["perc_tip"], kde=True)
plt.title("Distribuição da % de Gorjeta")
plt.show()

# Boxplot por dia
sns.boxplot(data=df, x="day", y="perc_tip")
plt.title("Distribuição da % de Gorjeta por Dia")
plt.show()

# Correlação
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlação entre variáveis numéricas")
plt.show()
```
