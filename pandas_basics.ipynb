```python
import pandas as pd

# Criar DataFrame
dados = {
    "Dia": ["Seg","Ter","Qua","Qui","Sex"],
    "Horas de Estudo": [2,3,1,4,2],
    "Produtividade": [6,7,5,8,6]
}
df = pd.DataFrame(dados)

# Exibir DataFrame centralizado (estilo)
df_styled = df.style.set_table_styles([
    {'selector': 'th', 'props': [('text-align', 'center')]},
    {'selector': 'td', 'props': [('text-align', 'center')]}
])

display(df_styled)  # Mostra a tabela formatada (Jupyter/VSCode)

# Estatísticas
print("\nEstatísticas básicas:")
print(df.describe())
```
